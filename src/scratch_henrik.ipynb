```python
import plotting
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon, Circle
```

```python
Theta = 2*np.pi/360*150
scale = 0.5
factor = 1

def triangle(x,y, phi):
    ax = (x+factor*scale*np.cos(phi))
    ay = (y+factor*scale*np.sin(phi))

    bx = (x+scale*np.cos(phi+Theta))
    by = (y+scale*np.sin(phi+Theta))

    cx = (x+scale*np.cos(phi-Theta))
    cy = (y+scale*np.sin(phi-Theta))

    triangle = np.array([[ax, ay], [bx, by], [cx, cy]])
    print(triangle)
    return triangle

alpha = np.pi/4
def polygon(datapoint = (0,0,alpha)):
    x, y, phi = datapoint
    return Polygon(triangle(x,y,phi,), facecolor = 'green', alpha=0.5)

p = polygon()
c = Circle((0,0),scale)

fig,ax = plt.subplots()

ax.add_patch(c)
ax.add_patch(p)

ax.axis('equal')
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)

plt.grid()
plt.show()
```
